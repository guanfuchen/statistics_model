```python
#coding:utf-8
import torch
import numpy as np
from torch.autograd import Variable
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# torch.linspace?
# F.sigmoid?
# plt.title?
# plt.axhline?
# F.leaky_relu?
# F.prelu?
```

```python
x_tensor = torch.linspace(-5, 5, 1000)
```

```python
x_variable = Variable(x_tensor)
```

```python
x_np = x_tensor.numpy()
```

```python
y_variable = F.sigmoid(x_tensor)
```

```python
y_tensor = y_variable.data
y_np = y_tensor.numpy()
```

```python
plt.plot(x_np, y_np)
plt.axhline(0, color='r', linestyle='--')
plt.title('Sigmoid function')
```

```python
# ReLU函数
y_variable = F.relu(x_tensor)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('ReLU function')
```

```python
# Tanh双曲正切函数
y_variable = F.tanh(x_tensor)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('Tanh function')
```

```python
# 反正切函数值域在[-pi/2, pi/2]
y_tensor = torch.atan(x_tensor)
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('arctan function')
```

```python
# LReLU双曲正切函数
y_variable = F.leaky_relu(x_tensor)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('Leaky ReLU function')
```

```python
# ELU函数
y_variable = F.elu(x_tensor)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('ELU function')
```

```python
# softplus函数
y_variable = F.softplus(x_tensor)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('softplus function')
```

```python
# softmax函数
y_variable = F.softmax(x_tensor)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('softmax function')
```

```python
# softmin函数
y_variable = F.softmin(x_variable)
y_tensor = y_variable.data
y_np = y_tensor.numpy()
plt.plot(x_np, y_np)
# plt.axvline(0, color='r', linestyle='--')
plt.axhline(0, color='r', linestyle='--')
plt.title('softmin function')
```
